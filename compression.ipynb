# Lossless Data Compression

## Lempel-Ziv Algorithm

Uses references to previously encoded data.

<http://en.wikipedia.org/wiki/LZ77_and_LZ78>

```python
ESCAPE = 255
MIN_REFERENCE_LEN = 4
MAX_REFERENCE_LEN = 254

def pack(data):
    output = bytearray()
    i = 0
    n = len(data)
    
    while i < n:
        #print(data[i:])
        
        tail = data[i-MAX_REFERENCE_LEN:i]
        length, offset = None, None
        for k in range(MIN_REFERENCE_LEN, min(n-i, MAX_REFERENCE_LEN)+1):
            try:
                j = tail.index(data[i:i+k])
                length = k
                offset = len(tail) - j
                #print("found possible reference", data[i-offset:i-offset+length])
            except ValueError:
                break
        
        if offset is None:
            x = data[i]
            i += 1
            
            output.append(x)
            if x == ESCAPE:
                output.append(x)
        else:
            print("reference for", data[i-offset:i-offset+length])
            i += length
            
            output.append(ESCAPE)
            output.append(offset)
            output.append(length)
    
    print("length:", n, "->", len(output), "saved %.2f %%" % (100*(1-(len(output)/n))))
    return bytes(output)

def unpack(data):
    it = iter(data)
    output = bytearray()
    
    try:
        while True:
            #print(output)
            
            x = next(it)
            if x == ESCAPE:
                y = next(it)
                if y == ESCAPE:
                    output.append(x)
                else:
                    offset, length = y, next(it)
                    print("duplicating offset =", offset, "length =", length, "data =", output[-offset:][:length])
                    output.extend(output[-offset:][:length])
            else:
                output.append(x)
    except StopIteration:
        pass
    
    return bytes(output)



text = "The true tautologic definition says that the true tautologic definition is true.".encode("utf-8")
print(">>>", text)
enc = pack(text)
print(">>>", enc)
dec = unpack(enc)
print(">>>", dec)
```

## Huffman Coding

Entropy-optimal encoding of input symbols into bits (variable length encoding).

<http://en.wikipedia.org/wiki/Huffman_coding>

Note: code tree would be stored along with encoded data (thus increasing file size a bit).

```python
from collections import Counter, namedtuple
import heapq

EOF = "EOF"

Node = namedtuple("Node", "value left right")

def make_tree(xs):
    heap = [(freq, i, Node(x, None, None)) for i, (x, freq) in enumerate(xs)]
    heapq.heapify(heap)
    
    while len(heap) >= 2:
        fa, i, a = heapq.heappop(heap)
        fb, j, b = heapq.heappop(heap)
        fab = fa + fb
        ab = Node(None, a, b)
        e = fab, i, ab
        heapq.heappush(heap, e)
    
    return heap.pop()[2]

def make_table(tree, table, prefix=()):
    if tree is not None:
        if tree.value is not None:
            table[tree.value] = prefix
        if tree.left is not None:
            make_table(tree.left, table, prefix+(0,))
        if tree.right is not None:
            make_table(tree.right, table, prefix+(1,))
           
            
def pack_bits(data):
    ctr = Counter(data) | Counter([EOF])
    tree = make_tree(ctr.most_common())
    table = {}
    make_table(tree, table)
    
    def gen():
        for x in data:
            print(chr(x),"->", table[x])
            yield from table[x]
        yield from table[EOF]
        print("EOF","->",table[EOF])
    
    return tree, gen()


def unpack_bits(root, bits):
    node = root
    
    for x in bits:
        if node.value is not None:
            if node.value == EOF:
                return
            else:
                #print("decoded", chr(node.value))
                yield node.value
                node = root

        if x == 0:
            node = node.left
        else:
            node = node.right


def read_bits(data):
    for x in data:
        for i in range(8):
            yield (x >> i) & 1

def concat_bits(bits):
    x = 0
    
    for i, b in enumerate(bits):
        x |= b << (i%8)
        
        if i % 8 == 7:
            yield x
            x = 0
    
    yield x
            
text = b"Hello world!"
print(">>>", text)

tree, bits = pack_bits(text)
data = bytes(concat_bits(bits))
print(">>>", len(data), "bytes")

bits = read_bits(data)
output = unpack_bits(tree, bits)
output = bytes(output)
print(">>>", output)
print(">>> compressed text size:", 100 * len(data) / len(text), "%")
```

## Combination of LZ77 and Huffman

```python
def combined_pack(data):
    lz_output = pack(data)
    huffman_tree, huffman_bits = pack_bits(lz_output)
    return huffman_tree, bytes(concat_bits(huffman_bits))

def combined_unpack(tree, data):
    huffmann_bits = read_bits(data)
    lz_input = bytes(unpack_bits(tree, huffmann_bits))
    return unpack(lz_input)


text = "The true tautologic definition says that the true tautologic definition is true.".encode("utf-8")
huffman = bytes(concat_bits(pack_bits(text)[1]))
lz77 = pack(text)
combined = combined_pack(text)[1]

print("----------------------")
print("Input data:", len(text), "bytes")
print("Huffman:", len(huffman), "bytes")
print("LZ77:", len(lz77), "bytes")
print("LZ77 -> Huffman:", len(combined), "bytes")
```
