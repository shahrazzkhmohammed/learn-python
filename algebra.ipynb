```python
from collections import namedtuple

Variable = namedtuple("Variable", "name")
Constant = namedtuple("Constant", "value")
BinOp = namedtuple("BinOp", "op left right")

OPERATIONS = {"+": lambda a, b: a + b, "-": lambda a, b: a - b, "*": lambda a, b: a*b, "/": lambda a, b: a/b}

def postfix2expr(text):
    stack = []
    for token in text.split():
        expr = None
        if token in "+-*/":
            right, left = stack.pop(), stack.pop()
            expr = BinOp(token, left, right)
        elif token.isnumeric():
            expr = Constant(int(token))
        else:
            expr = Variable(token)
        stack.append(expr)
    
    return stack.pop()

def expr2infix(expr):
    if isinstance(expr, Variable):
        return expr.name
    elif isinstance(expr, Constant):
        return str(expr.value)
    elif isinstance(expr, BinOp):
        return "(%s)" % " ".join([expr2infix(expr.left), expr.op, expr2infix(expr.right)])
    else:
        raise TypeError()

def isconstant(expr):
    if isinstance(expr, Variable):
        return False
    elif isinstance(expr, Constant):
        return True
    elif isinstance(expr, BinOp):
        return isconstant(expr.left) and isconstant(expr.right)
    else:
        raise TypeError()

def substitute_value(expr, **variables):
    if isinstance(expr, Variable):
        if expr.name in variables:
            return Constant(variables[expr.name])
        else:
            return expr
    elif isinstance(expr, Constant):
        return expr
    elif isinstance(expr, BinOp):
        return BinOp(expr.op, substitute_value(expr.left, **variables), substitute_value(expr.right, **variables))
    else:
        raise TypeError()

def fold_constants(expr):
    if isinstance(expr, Variable):
        return expr
    elif isinstance(expr, Constant):
        return expr
    elif isinstance(expr, BinOp):
        left, right = fold_constants(expr.left), fold_constants(expr.right)
        if isinstance(left, Constant) and isinstance(right, Constant):
            value = OPERATIONS[expr.op](left.value, right.value)
            return Constant(value)
        else:
            return BinOp(expr.op, left, right)
    else:
        raise TypeError()
        
data = "a a * b b * c c * + +"
expr = postfix2expr(data)
infix = expr2infix(expr)
print(data, expr, infix, sep="\n")
subs=dict(a=3, b=0)
expr2 = substitute_value(expr, **subs)
infix2 = expr2infix(expr2)
print("substituting %r" % subs, expr2, infix2, sep="\n")
expr3 = fold_constants(expr2)
infix3 = expr2infix(expr3)
print(expr3, infix3, sep="\n")
```
